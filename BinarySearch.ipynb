# **Binary Search in a Array**

```python
## Binary Search Algorithm
## function definition
def binarySearch(arr, x, i, j):
  while i <= j:
    mid = i + (j-i) // 2

    if arr[mid] == x:
      return mid

    elif arr[mid] < x:
      ## right side of the mid
      ## search space - mid + 1 to j
      i = mid + 1

    else:
      ## left side of the mid
      ## search space - i to mid - 1
      j = mid -1
  ## Searching element is not present in array
  return -1

## Driver code
arr = [20, 30, 40, 50, 60, 70, 80, 90]
x = 54
i = 0
j = len(arr) - 1

result = binarySearch(arr, x, i, j)
print ("Searchig element is present at", result)
```
