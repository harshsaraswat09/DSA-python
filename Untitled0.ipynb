```python
## Implementation of insertion sort
## Time complexity: O(n^2)
def insertionSort(arr):
  for i in range(len(arr)):
    j = i - 1
    key = arr[i]

    while j >= 0 and key < arr[j]:
      arr[j + 1] = arr[j]
      j = j - 1
    arr[j + 1] = key

  return arr


## Driver code
arr = [75, 90, 100, 95, 85, 80]
result = insertionSort(arr)
print(result)
```
